```python
list_feedback=[]
list_target=[]
with open('en-development.txt') as develop:
    for i in develop:
        #print(i)
        feedback=i.strip().lower().split()
        #print(temp_feedback)
        if len(feedback) >=1:
            target=feedback[-1]
            del(feedback[0])
            del(feedback[-1])
            list_feedback.append(feedback)
            list_target.append(target)
```

```python
list_target
```

```python
stopwords=[]
with open('stopwords.txt') as stops:
    for i in stops:
        stopwords.append(i.strip())
        print(i)
```

```python
temp_feed=[]
for i in list_feedback:
    refined_data=[]
    for j in i:
        if  j not in stopwords:
            refined_data.append(j)
    temp_feed.append(refined_data)
```

```python
temp_feed
```

```python
from collections import defaultdict
#calculate the frequency of the words
fre=defaultdict(int)
for i in temp_feed:
     for j in i:
        fre[j]=fre[j]+1

refined_feedback=[]
for i in temp_feed:
        temp_feedb=[]
        for j in i:
                if fre[j]>1:
                    temp_feedb.append(j)
        refined_feedback.append(temp_feedb)
len(refined_feedback)
```

```python
list_target
```

```python
count=0
for i in refined_feedback:
    if len(i)==0:
        refined_feedback.remove(i)
        list_target.pop(count)
    count=count+1
```
